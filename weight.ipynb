# DEEP LEARNING MODEL

````python
#normalize (z score)
def norm(x):
  return(x- train_stat['mean'])/train_stat['std'] #keras.layers.Normalization(mean=train_stat['mean'],varicance=train_stat["varicance"])
normed_train_data= norm(train_data)
normed_test_data= norm(test_data)
````

````python
model= tf.keras.Sequential([
                            #keras.layers.Normalization(mean=train_stat['mean'],varicance=train_stat["varicance"])
                             tf.keras.layers.Dense(48,activation="relu", input_shape=(48,)),
                             tf.keras.layers.Dense(196, activation="relu"),
                             tf.keras.layers.Dense(196, activation="relu"),
                            tf.keras.layers.Dense(48, activation='relu'),
                             tf.keras.layers.Dense(1)
 ])
````

````python
model.summary()
````

````python
model.compile(loss='mse',
              optimizer='RMSprop',
              metrics=['mae', 'mse'])
````

````python
ex= normed_train_data[:48]
ex_result= model.predict(ex)
np.argmin(ex_result)
````

````python
import keras
# check for model training
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if(epoch%100==0):print(' ')
    print('-', end="")
# to help to avoid overfitting
early_stop=keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)
history= model.fit(normed_train_data, train_labels,
                   epochs=1000,
                   validation_split=0.2,
                   verbose=0,
                   callbacks=[PrintDot(), early_stop])
````

````python
hist =pd.DataFrame(history.history)
hist
````

````python
import matplotlib.pyplot as plt
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch
  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('weight')
  plt.plot(hist['epoch'], hist['mae'], label='train_error')
  plt.plot(hist['epoch'], hist['val_mae'],label='val Error')
  #plt.legend()
  plt.ylim([0,200])
  plt.figure()
  plt.xlabel('epoch')
  plt.ylabel('weight')
  plt.plot(hist['epoch'], hist['mse'], label='train_error')
  plt.plot(hist['epoch'], hist['val_mse'], label='val error')
  plt.legend()
  plt.ylim([0,15000])
plot_history(history)
````

````python
test_pre =model.predict(normed_test_data)
plt.scatter(test_labels, test_pre)
plt.xlabel('true label')
plt.ylabel('prediction label')
plt.axis('equal')
plt.axis('square')
plt.xlim([0, plt.xlim()[1]])
plt.ylim([0, plt.ylim()[1]])
_ =plt.plot([-100,300], [-100,300])
````
